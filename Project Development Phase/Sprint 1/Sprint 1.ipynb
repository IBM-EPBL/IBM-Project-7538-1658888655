```python
plt.figure(figsize =(14,5))
plt.subplot(1,2,1)
sns.countplot(df['Salary'] ,color ='b' ,hue =df['YearsExperience'])
plt.title('Attrition by Gender')
plt.subplot(1,2,2)
```

```python
plt.figure(figsize =(16 ,4))
plt.subplot(1,3,1)
sns.distplot(df['Salary'] ,color ='m')
plt.title('graph')
plt.subplot(1 , 3 ,2)
sns.stripplot(x = 'Gender' ,y = 'Age' ,data = df ,palette="Set2")
plt.title('Gender vs Age')
plt.subplot(1,3,3)
sns.countplot('Gender' ,data = df ,color ='c')
plt.title('Gender')
plt.tight_layout()
```

```python
plt.figure(figsize = (14 , 13))
plt.subplot(2 ,1,1)
sns.countplot(y= 'Gender' ,data = df ,palette='winter_r')
plt.title('Gender')
plt.subplot(2,1,2)
sns.countplot(y= 'Age' ,data = df ,palette='winter_r'  ,hue =df['Age'])
```

```python
plt.figure(figsize =(14,5))
plt.subplot(1,2,1)
sns.countplot('YearsExperience' ,data = df ,hue ='Age' ,palette='gist_rainbow')
plt.subplot(1,2,2)
plt.pie(df['YearsExperience'].value_counts() ,autopct ='%.2f' ,colors = ['r' ,'c' ,'g'],labels =['Research & Development','Sales', 'Human Resources'] ,explode =[0 ,0.1,0])
```

```python
#HANDLING CATEGORICAL OUTPUT VARIABLE
df['Age'].replace({'Yes':1 ,'No':0} ,inplace = True)
df['Age'].head()
```

```python
plt.figure(figsize =(14 ,10))
plt.subplot(2,2,1)
sns.countplot(df['YearsExperience'] ,hue =df['Age'] ,palette='Accent_r') 
plt.subplot(2,2,2)
sns.countplot(df['Gender'] ,hue =df['Age'] ,palette='Accent')
plt.subplot(2,2,3)
sns.countplot(df['Gender'] ,hue =df['Age'] ,palette='brg_r')
plt.subplot(2,2,4)
sns.countplot(df['Salary'] ,hue =df['Age'] ,palette='twilight_r')
```

```python
plt.figure(figsize =(20 ,8))
sns.boxplot(x ='YearsExperience', y = 'Salary' ,data = df ,hue ='Age' ,color ='red')
```
